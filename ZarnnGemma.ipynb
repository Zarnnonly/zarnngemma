## Run the model (for the time being, just simple output from cell, might add openai compatible later)

```python
from vllm import LLM, SamplingParams
from transformers import AutoTokenizer

model_name = "google/gemma-4-31B-it" 

llm = LLM(
    model=model_name,
    dtype="bfloat16", # Change if u want to use quantized ver
    gpu_memory_utilization=0.9, # GPU memory allocation, dont change if u dont know what u're doing
    max_model_len=4096, # Context size in token, dont change if u dont know what u're doing
    enforce_eager=True,
)

tokenizer = AutoTokenizer.from_pretrained(model_name) # Tokenizer setup

# Self explanatory things
sampling_params = SamplingParams(
    temperature=0.9,
    top_p=0.9,
    max_tokens=2048,
)
```

```python
# Run and change the prompt here

messages = [
    {"role": "user", "content": "Jelaskan kenapa beberapa dosen sangat suka cuma ngasih presentasi terus sibuk sendiri"}
]

prompt = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True,
)

outputs = llm.generate([prompt], sampling_params)
print(outputs[0].outputs[0].text)
```
